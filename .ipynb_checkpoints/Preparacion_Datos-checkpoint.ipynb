# Librerias

```python
# Para trabajar los datos
import pandas as pd
import numpy as np

# Para visualizar los datos
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Type (Muestra/ for "muestra" and Completos/ for "completos")
tipo = "Completos/"
#tipo = "Muestra/"
```

```python
#Dimensions
d_city         = tipo + "dimension_city"
d_customer     = tipo + "dimension_customer"
d_date         = tipo + "dimension_date"
d_employee     = tipo + "dimension_employee"
d_stock_item   = tipo + "dimension_stock_item"
#Fact table
f_order        = tipo + "fact_order"
#Dataframes
df_city         = pd.read_csv(f'data/{d_city}.csv', sep=',')
df_customer     = pd.read_csv(f'data/{d_customer}.csv', sep=',')
df_date         = pd.read_csv(f'data/{d_date}.csv', sep=',')
df_employee     = pd.read_csv(f'data/{d_employee}.csv', sep=',')
df_stock_item   = pd.read_csv(f'data/{d_stock_item}.csv', sep=',')
df_order        = pd.read_csv(f'data/{f_order}.csv', sep=',')
```

## Revisamos la dimensión stockitem

```python
df_stock_item.head()
```

```python
df_stock_item.info()
```

Tengo miedo de quitar una columna entera porque a lo mejor y luego es útil, so vamos a llenar las columnas color, brand y size_val como unknown.

```python
df_stock_item["Color"] = df_stock_item["Color"].apply(lambda x : "Unknown" if pd.isna(x) else x)
df_stock_item["Brand"] = df_stock_item["Brand"].apply(lambda x : "Unknown" if pd.isna(x) else x)
df_stock_item["Size_val"] = df_stock_item["Size_val"].apply(lambda x : "Unknown" if pd.isna(x) else x)
```

Ahora si, removemos las filas con elementos nulos.

```python
len_o = len(df_stock_item)
df_stock_item.dropna(inplace = True)
```

Podemos ver las filas eliminadas

```python
len_f = len(df_stock_item)
len_o-len_f
```

Convertimos Tax_Rate, Unit_Price, Recommended_Retail_Price, Typical_Weight_Per_Unit a float

```python
#Quitamos las ','
df_stock_item["Tax_Rate"] = df_stock_item["Tax_Rate"].apply(lambda x: x.replace(",","."))
df_stock_item["Unit_Price"] = df_stock_item["Unit_Price"].apply(lambda x: x.replace(",","."))
df_stock_item["Recommended_Retail_Price"] = df_stock_item["Recommended_Retail_Price"].apply(lambda x: x.replace(",","."))
df_stock_item["Typical_Weight_Per_Unit"] = df_stock_item["Typical_Weight_Per_Unit"].apply(lambda x: x.replace(",","."))
#Convertimos a float
df_stock_item["Tax_Rate"] = df_stock_item["Tax_Rate"].astype(np.float64)
df_stock_item["Unit_Price"] = df_stock_item["Unit_Price"].astype(np.float64)
df_stock_item["Recommended_Retail_Price"] = df_stock_item["Recommended_Retail_Price"].astype(np.float64)
df_stock_item["Typical_Weight_Per_Unit"] = df_stock_item["Typical_Weight_Per_Unit"].astype(np.float64)
```

Tenemos que remover los registros de la tabla de hechos que hagan referencia a algún elemento removido de city

```python
len_o = len(df_order)
if tipo == "Completos/":
    df_order = df_order[df_order.Stock_Item_Key.isin(df_stock_item.Stock_Item_Key)]
else:
    df_order = df_order[df_order.stock_item_key.isin(df_stock_item.Stock_Item_Key)]
df_order
```

Podemos ver las filas eliminadas

```python
len_f = len(df_order)
len_o-len_f
```

## Revisamos cómo queda la tabla de hechos

```python
df_order.head()
```

Convertimos Unit_Price, Tax_Rate, Total_Excluding_Tax, Tax_Amount, Total_Including_Tax a float

```python
#Quitamos las ','
df_order["Unit_Price"] = df_order["Unit_Price"].apply(lambda x: x.replace(",","."))
df_order["Tax_Rate"] = df_order["Tax_Rate"].apply(lambda x: x.replace(",","."))
df_order["Total_Excluding_Tax"] = df_order["Total_Excluding_Tax"].apply(lambda x: x.replace(",","."))
df_order["Tax_Amount"] = df_order["Tax_Amount"].apply(lambda x: x.replace(",","."))
df_order["Total_Including_Tax"] = df_order["Total_Including_Tax"].apply(lambda x: x.replace(",","."))
#Convertimos a float
df_order["Unit_Price"] = df_order["Unit_Price"].astype(np.float64)
df_order["Tax_Rate"] = df_order["Tax_Rate"].astype(np.float64)
df_order["Total_Excluding_Tax"] = df_order["Total_Excluding_Tax"].astype(np.float64)
df_order["Tax_Amount"] = df_order["Tax_Amount"].astype(np.float64)
df_order["Total_Including_Tax"] = df_order["Total_Including_Tax"].astype(np.float64)
```
